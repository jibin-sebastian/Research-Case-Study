```python
import pandas as pd
import numpy as np
import seaborn as sns
import os
import matplotlib.pyplot as plt
import plotly.express as px
from datetime import datetime
```

```python
### Set your path to the folder containing the .csv files
PATH = '/content/drive/MyDrive/RCS/smartphone_sct/' # Use your path (copy sct folder path from colab left pane)
#print(PATH+'10e5b7cd.csv')

### Fetch all files in path
fileNames = os.listdir(PATH)

### Filter file name list for files ending with .csv
fileNames = [file for file in fileNames if '.csv' in file]
#print(fileNames)
count=0
def time(df):
  timeCol=df['time'].tolist()
  time=[]
  for i in timeCol:
    p=str(i)
    tm=p[0:(len(p)-3)]
    ts = int(tm)
    #ti=datetime.utcfromtimestamp(ts).strftime('%Y-%m-%d %H:%M:%S')
    ti=datetime.utcfromtimestamp(ts).strftime('%H:%M:%S')
    #time.append(ti+':'+(p[-3:]))
    time.append(ti)
  return time


### Loop over all files
for file in fileNames:
  count=count+1
  c=str(count)
  ### Read .csv file and append to list
  data = pd.read_csv(PATH + file,delimiter=';')
  df=data.loc[(data['type_profile'] == 'basic') & (data['type_category']== 'acceleration')]
  result_df=time(df)
  df1 = pd.DataFrame (result_df, columns = ['time'])
  data['time']=df1['time']
  tit="Time vs Value "+file+'-- '+c
  fig = px.scatter(data, x="time", y="value",title=tit)
  fig.show()
```
